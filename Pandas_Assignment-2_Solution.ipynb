```python
def text_len(DataFrame):
    Word_count = []
    text_list = list(DataFrame["Text"])
    for text in text_list:
        len_text = len(text)
        Word_count.append(len_text)
    
    DataFrame['Word_count'] =  Word_count 
    return DataFrame
```

#### Answer_5

##### DataFrame.size() and DataFrame.shape() are both methods in Pandas library in Python used to get information about the dimensions of a DataFrame object, but they provide different information.

##### DataFrame.size() returns the total number of elements in the DataFrame, which is equal to the product of the number of rows and the number of columns. In other words, DataFrame.size() returns the total number of cells in the DataFrame.

##### DataFrame.shape() returns a tuple representing the dimensions of the DataFrame. The tuple contains two values: the number of rows and the number of columns. For example, if a DataFrame has 5 rows and 3 columns, DataFrame.shape() will return (5, 3).

```python
df9 = pd.DataFrame({
    'Country' : ['Afghan','China','Bangladesh','Japan','Pakistan','UAE','USA'],
    'Value'   : [0.95,11.86,0.79,0.61,0.37,22.53,82.73]
})
```

```python
def statistics(DataFrame):
    Mean = DataFrame['Value'].mean()
    Median =  DataFrame['Value'].median()
    Standard_Deviation = DataFrame['Value'].std()
    return f"Mean is {Mean} , Median is {Median} , Standard Deviation is {Standard_Deviation}"
```

```python
statistics(df9)
```
